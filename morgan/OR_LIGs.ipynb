```python

# Calculate distances using different metrics for each molecule in fingerprint file 1 against filtered molecules in fingerprint file 2
dist_hamming = pairwise_distances(df_filtered.drop('SMILES', axis=1), df=df_1.drop('Fingerprint', axis=1), metric='hamming')
dist_cosine = pairwise_distances(df_filtered.drop('SMILES', axis=1), df=df_1.drop('Fingerprint', axis=1), metric='cosine')
dist_tanimoto = pairwise_distances(df_filtered.drop('SMILES', axis=1), df=df_1.drop('Fingerprint', axis=1), metric='jaccard')
dist_hyperbolic = pairwise_distances(df_filtered.drop('SMILES', axis=1), df=df_1.drop('Fingerprint', axis=1), metric='hyperbolic')
dist_euclidean = pairwise_distances(df_filtered.drop('SMILES', axis=1), df=df_1.drop('Fingerprint', axis=1), metric='euclidean')
```
